```python
import pandas as pd
import pymongo
from pymongo import MongoClient
from datetime import datetime
```

```python
client = MongoClient("mongodb://localhost:27017/")
```

```python
db = client["electric_power_consumption"]
collection = db["cleaned_data"]
```

```python
df = pd.read_csv('cleaned_data.csv')  
df
```

#### single record with the highest Global_active_power

```python
peak_record = collection.find_one(
    sort=[("Global_active_power", -1)]  
)
peak_record
```

#### timestamp and the power value

```python
print("Timestamp:", peak_record.get("Timestamp"))
print("Global_active_power:", peak_record.get("Global_active_power"))
```

#### Calculate power consumption for each month of the year 2008.

```python
df['Global_active_power'] = pd.to_numeric(df['Global_active_power'])
df['particular_year'] = (df['Date'] >= '2008-01-01') & (df['Date'] < '2009-01-01')
df['Particular_month']=particular_year['Date'].dt.to_period('M')
consumption = df.groupby('Particular_month')['Global_active_power'].sum().reset_index()
consumption
```
